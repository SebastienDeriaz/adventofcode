```python
N = 10
values = [1,2,3]

def rc(n : int, values : list[int]):
    d = {v : 0 for v in values}
    v = values[0]
    if len(values) == 1:
        d[v] += n
        yield d
        return None
    
    for i in range(0, n+1):
        for output in rc(n-i, values[1:]):
            di = d.copy()
            di[v] += i
            for button, counts in output.items():
                di[button] += counts
            yield  di



for x in rc(N, values):
    print(x)
```

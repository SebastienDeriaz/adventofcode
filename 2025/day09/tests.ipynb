```python
import numpy as np
import matplotlib.pyplot as plt



a, c = np.random.uniform(-10, 10, (2,2))

b = np.array([a[0], c[1]])
d = np.array([c[0], a[1]])


s = np.random.uniform(-10, 10, (2))
e = np.random.uniform(-10, 10, (2))
k = np.random.randint(0, 2)
e[k] = s[k]
R = np.stack([a, b, c, d, a])


plt.figure()
plt.plot([s[0], e[0]], [s[1], e[1]])
plt.plot(R[:, 0], R[:, 1])
plt.scatter(*a)
plt.scatter(*c)
plt.xlim(-10,10)
plt.ylim(-10,10)
plt.show()
```

# Advent of Code - Day 6

- total race time : $T$
- time spent pressing the button : $t_b$
- initial speed : $v_0$

Initial speed
$$k_v = 1\text{mm}/\text{ms}Â²$$

$$v_0 = t_b \cdot k_v$$

Distance travelled

$$d = (T-t_b)*v_0

If we want to know what time the button is pressed $t_b$ is necessary go beyond the goal :

$$d > d_g$$
$$(T-t_b)\cdot v_0 > d_g$$
$$(T-t_b)\cdot(t_b\cdot k_v) > d_g$$
$$t^2_b - Tt_b + \frac{d_g}{k_v} < 0 $$

We look for the roots

$$t^2_b - Tt_b + \frac{d_g}{k_v} = 0 $$

$${t_b}_1, {t_b}_2 = \frac{-b\pm\sqrt{b^2-4ac}}{2a}=\frac{T\pm\sqrt{T^2-4\frac{d_g}{k_v}}}{2}$$

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
kv = 1

def distance_travelled(T, t_b):
    v0 = kv * t_b
    return (T-t_b)*v0



def roots(T, d_g, k_v):
    x = np.sqrt(T**2 - 4*d_g/k_v)
    return (T - x)/2, (T + x)/2


d_g = 200
T = 30
k_v = 1

d = distance_travelled(d_g, np.arange(d_g+1))

t_b_1, t_b_2 = roots(T, d_g, k_v)
print(t_b_1, t_b_2)
n_ways = np.floor(t_b_2) - np.ceil(t_b_1) + 1
print(n_ways)
plt.plot(d, '.-')
plt.grid()
plt.show()
```

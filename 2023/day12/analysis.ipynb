```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

df = pd.read_csv('values.csv', names=['line', 'n1', 'n2', 'n3', 'n4'], index_col=0)

df
```

```python
plt.figure(figsize=(8,8))
for i, row in df.iterrows():
    plt.semilogy(row, '.-', label=f'{i}')
plt.legend()
plt.grid()
plt.show()
```

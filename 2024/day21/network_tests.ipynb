```python
import networkx as nx
import numpy as np
import matplotlib.pyplot as plt


NUMERIC_KEYPAD = np.array([
    ['7', '8', '9'],
    ['4', '5', '6'],
    ['1', '2', '3'],
    [None, '0', 'A']
])

DIRECTIONAL_KEYPAD = np.array([
    [None, '^', 'A'],
    ['<', 'v', '>']
])




def make_network(keypad : np.ndarray) -> nx.DiGraph():
    G = nx.DiGraph()
    for x in keypad.flatten():
        if x is not None:
            G.add_node(x)
    
    
    for r, c in np.ndindex(keypad.shape):
        for dr, dc, symbol in [
            (0, 1, '>'),
            (0, -1, '<'),
            (1, 0, 'v'),
            (-1, 0, '^')
            ]:
            r2 = r + dr
            c2 = c + dc
            if 0 <= r2 < keypad.shape[0] and 0 <= c2 < keypad.shape[1]:
                if keypad[r, c] is not None and keypad[r2, c2] is not None:
                    # Valid
                    G.add_edge(keypad[r, c], keypad[r2, c2], key=symbol, weight=1)

    return G
```

```python
G = make_network(NUMERIC_KEYPAD)


nx.draw_networkx(G)
plt.show()


nx.astar_path(G, 'A', '7')
```
